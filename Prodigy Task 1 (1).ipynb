```python
#Check & plot 10 countries with maximum population in the year 2001 
```

```python
max_10_pop_2001 = df.sort_values(by='2001', ascending=False).head(10)
```

```python
plt.figure(figsize=(6,3))
plt.bar(max_10_pop_2001['Country Name'], max_10_pop_2001['2001'], color='red')
plt.title('Top 10 Countries by Population in the Year 2001')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Population')
```

#In same way we can check for any year top 10 population country name 

```python
#Check & plot 10 countries with minimum population in the year 2002
```

```python
min_10_pop_2002 = df.sort_values(by='2002', ascending=True).head(10)
```

```python
plt.figure(figsize=(6,3))
plt.bar(max_10_pop_2001['Country Name'], min_10_pop_2002['2002'], color='purple')
plt.title('Top 10 Countries with minmum Population in the Year 2001')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Population')
```

#In same way we can check for any year top 10 with minmum population of country name 

```python
#Check and plot which year have maximum polpluation
```

```python
max_pop_year = df.loc[:, '2001':'2022'].idxmax(axis=1)
```

```python
year_counts = max_pop_year.value_counts()

# Plot the year with maximum population for each country
plt.figure(figsize=(6, 3))
year_counts.plot(kind='bar', color='green')
plt.title('Year with Maximum Population for Each Country')
plt.xlabel('Year')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot the year wise mean population for each year
```
